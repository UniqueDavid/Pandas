```python
def hundredth_row(column):
    hundredth_item = column.loc[99]
    return hundredth_item

hundred_row = titanic_survival.apply(hundredth_row)
print hundred_row
```

```python
def not_null_count(column):
    column_null = pd.isnull(column)
    null = column[column_null]
    return len(null)
column_null_count = titanic_survival.apply(not_null_count)
print column_null_count
```

```python
def which_class(row):
    pclass = row['Pclass']
    if pd.isnull(pclass):
        return "Unknown"
    elif pclass == 1:
        return "First Class"
    elif pclass == 2:
        return "Second Class"
    elif pclass == 3:
        return "Third Class"
classes = titanic_survival.apply(which_class, axis=1)
print classes
```

```python
def generate_age_label(row):
    age = row["Age"]
    if pd.isnull(age):
        return "unknown"
    elif age < 18:
        return "minior"
    else:
        return "adult"

age_labels = titanic_survival.apply(generate_age_label, axis = 1)
print age_labels
```
